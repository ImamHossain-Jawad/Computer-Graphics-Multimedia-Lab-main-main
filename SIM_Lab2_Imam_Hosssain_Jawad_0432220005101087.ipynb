# Student's Assignment 1: Basic Matrix Operations

```python
import numpy as np

# Create two 3×3 matrices with random integers between 1 and 10
A = np.random.randint(1, 17, size=(3, 3))
B = np.random.randint(1, 17, size=(3, 3))

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# Matrix Addition
add_result = A + B
print("\nAddition:\n", add_result)

# Matrix Subtraction
sub_result = A - B
print("\nSubtraction:\n", sub_result)

# Matrix Multiplication
mul_result = A @ B
print("\nMultiplication:\n", mul_result)

# Determinant of A
det_A = np.linalg.det(A)
print("\nDeterminant of A:", det_A)

# Inverse of A (if invertible)
if det_A != 0:
    inv_A = np.linalg.inv(A)
    print("\nInverse of A:\n", inv_A)
else:
    print("\nMatrix A is not invertible (det = 0)")

# Rank of A
rank_A = np.linalg.matrix_rank(A)
print("\nRank of A:", rank_A)
```

#### Necessary for Feature Selection in ML

### 1. **Determinant of a Matrix**:
The **determinant** is a special number that we calculate from a square matrix (a matrix with the same number of rows and columns). It can tell us a lot about the matrix. For example:
- If the **determinant is zero**, it means the matrix is **singular** (not invertible). This means we can't perform certain operations on it, like finding its inverse.
- If the **determinant is non-zero**, the matrix is **invertible**, meaning we can find an inverse matrix that "undoes" the effect of the original matrix.

### 2. **Inverse of a Matrix**:
The **inverse** of a matrix is another matrix that, when multiplied by the original matrix, gives the **identity matrix** (a special matrix that has 1s on the diagonal and 0s everywhere else).
- A matrix is only **invertible** if its determinant is not zero. If the determinant is zero, the matrix doesn’t have an inverse (it's "singular").
- The inverse of a matrix is useful when you want to "reverse" its operation, similar to how you would solve for a variable in an equation.

### 3. **Rank of a Matrix**:
The **rank** of a matrix tells you how many linearly independent rows or columns it has. A matrix's rank helps determine:
- How much "information" the matrix contains.
- If the matrix represents a system of equations, the rank tells you if there are enough independent equations to find a unique solution.
- The rank is the maximum number of linearly independent rows or columns in the matrix. For example, a matrix with full rank has all of its rows or columns providing unique information.



### 4. **Eigenvalues and Eigenvectors**:
- **Eigenvectors** are special directions or "paths." If you apply a matrix to an eigenvector, that vector doesn’t turn or twist. It just gets **stretched** or **shrunk**.
  - Imagine stretching a rubber band in a straight line. You can stretch it longer, but it still stays in the same direction. That’s like an eigenvector.
  
- **Eigenvalues** are the numbers that tell you how much the eigenvector gets stretched or shrunk.
  - If the eigenvalue is 2, the eigenvector gets **twice as long** (Twice data). If it’s 0.5, it gets **half as long** (less density of data).
  - Eigenvalues tell you **how much** stretching or shrinking happens. Used in image compression. Feature selection process like PCA.

In simpler terms:
- **Eigenvectors** are the directions that stay the same when the matrix is applied to them.
- **Eigenvalues** are how much those directions stretch or shrink.

Both are important for understanding things like image transformations in computer graphics, analyzing systems, or even predicting behavior in certain situations!

### 5. **Characteristic Polynomial**:
The **characteristic polynomial** is an equation derived from a matrix that helps us find its **eigenvalues**.
- The roots (solutions) of this polynomial are the **eigenvalues** of the matrix.
- It’s a useful tool for solving problems involving eigenvalues and understanding matrix properties.

1.enerate and Plot Two Random Vectors

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Part 1: Generate and Plot Two Random Vectors
vec1 = np.random.rand(15)
vec2 = np.random.rand(15)

plt.figure(figsize=(8, 4))
plt.plot(vec1, label='A', marker='o')
plt.plot(vec2, label='B', marker='s')
plt.title("Random Vectors Comparison")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

2.Generate and Visualize a Random Matrix:

3.Matrix Operations and Visualization:

```python
# Part 3: Matrix Operations and Visualization
M1 = np.random.rand(4, 4)
M2 = np.random.rand(4, 4)

# Addition
add_result = M1 + M2

# Subtraction
sub_result = M1 - M2

# Multiplication
mul_result = M1 @ M2

# Flatten each result for bar plot
def plot_matrix_bar(matrix, title):
    flat = matrix.flatten()
    plt.figure(figsize=(6, 4))
    plt.bar(range(len(flat)), flat)
    plt.title(title)
    plt.xlabel("Element Index")
    plt.ylabel("Value")
    plt.grid(True)
    plt.show()

plot_matrix_bar(add_result, "Matrix Addition Result")
plot_matrix_bar(sub_result, "Matrix Subtraction Result")
plot_matrix_bar(mul_result, "Matrix Multiplication Result")
```
